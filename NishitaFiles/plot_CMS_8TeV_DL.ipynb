```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as tick
import matplotlib as mpl
from matplotlib import cm
from matplotlib.colors import LogNorm
from math import *
%matplotlib inline
```

```python
tauvals = list()
Qratio_40 = list()
Qratio_30 = list()
Qratio_25 = list()
Qratio_20 = list()
```

```python
def read_file_Qratio(filename, Qratio_arr):
    for line in open(filename):
        vals = line.split()
        if (len(vals)) < 7: 
            break
        else:
            tauvals.append(float(vals[0]))
            Qratio_arr.append(float(vals[6]))
    return
```

```python
read_file_Qratio("tau-40.dat",Qratio_40)
read_file_Qratio("tau-30.dat",Qratio_30)
read_file_Qratio("tau-25.dat",Qratio_25)
read_file_Qratio("tau-20.dat",Qratio_20)
```
